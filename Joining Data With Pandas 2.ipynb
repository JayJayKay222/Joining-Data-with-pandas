```python

# Read the file
car_price = pd.read_csv('car Price', index_col=0)
car_price.head()
```

```python
# Merge the 2 tables in a sorted order
pd.merge_ordered(car_sales, car_price, on='ID',suffixes=['_sales','_price']).head()
```

```python

# Fill the Missing values with their Preceding row
pd.merge_ordered(car_sales, car_price, on='ID',suffixes=['_sales','_price'], fill_method='ffill').head()
```
