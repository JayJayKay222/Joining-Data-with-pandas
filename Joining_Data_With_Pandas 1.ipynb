```python
# SEMI JOIN
""" STEP 1 - Inner Join """
p = data.merge(x, on='title')
p.head()
```

```python

""" STEP 2 - Compare the 2 columns """
q = data[data['title'].isin(p['title'])]
q.head()
```

```python
# ANTI JOIN
""" STEP 1 - Return columns on the left table which is not in the right table Using left join """
s = data.merge(x, on='title', how='left', indicator=True)
s.head()
```

```python
""" STEP 2 - Subset the _merge column with left only rows """
b = s.loc[s['_merge'] == 'left_only', 'title']
b.head()
```

```python
# Concatenating DataFrames Vertically
pd.concat([l,x], ignore_index=True, sort=True).head()
```

```python

# Concatenating DataFrames with common columns
pd.concat([x,data], join='inner').head()
```

```python

# Validating the Integrity of your DataFrames
data.merge(x, on='title', how='outer', validate='one_to_many').head()
```
